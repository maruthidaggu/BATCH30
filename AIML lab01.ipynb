```python
people=['a','b','c','d']
age=[20,21,23,34]
weight=[44,47,48,49]
height=[160,167,180,178]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of the patients")
plt.ylabel("height in metres")
plt.xlabel("weight of the patients in kgs")
plt.show()
```

```python
plt.plot(age,height)
plt.title("Relationship between age and height of the patients")
plt.ylabel("height in metres")
plt.xlabel("age of the patients in cms")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.title("people's weight in kgs ",fontsize=16,fontstyle='italic')

plt.bar(x=people,height=weight,width=0.6,color='purple',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(IN KGS)",fontsize=15)
```
